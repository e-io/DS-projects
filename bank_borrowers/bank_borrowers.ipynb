--------

Посмотрим варианты идентификаторов образования. Напишем для удобства функцию variants_count, которая выводит данные по алфавиту, а не по количеству.

```python
def variants_count(column):
    print(f'Колонка `{column}`:')
    for i in sorted(df[column].unique().tolist()):
        print(f'{i}: {df[column][df[column] == i].count()}')
        
variants_count('education_id')
```

Их оказалось пять, а не два, как предполагалось ранее. Посмотрим сами названия вариантов образования.

```python
variants_count('education')
```

### Шаг 2.6. Категоризация дохода.

Категоризируем доход. Присвоим в зависимости от дохода категории A, B, C, D и E.

```python
def income_categ(a):
    a = a // 1_000
    if a < 30:
        return 'E'
    elif a < 50:
        return 'D'
    elif a < 200:
        return 'C'
    elif a < 1_000:
        return 'B'
    return 'A'

df['total_income_category'] = df['total_income'].agg(income_categ)
```

```python
def purpose_categ(a):
    if 'свадьб' in a:
        return 'проведение свадьбы'
    elif 'недвиж' in a or 'жиль' in a:
        return 'операции с недвижимостью'
    elif 'автомобил' in a:
        return 'операции с автомобилем'
    elif 'образов'in a:
        return 'получение образования'
    return 'Undefined category'
    
df['purpose_category'] = df['purpose'].agg(purpose_categ)
```

На данные самой левой колонки закрываем глаза, это имеется ввиду не состояние задолженности 0...5, а просто обычный индекс.

$$ p = \frac{b}{a + b} * 100 $$

```python
data_children_pivot
```
